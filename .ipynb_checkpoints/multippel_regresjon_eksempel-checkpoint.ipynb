Innlesning av serier fra SMG:

```python
import pytz
import matplotlib.pyplot as plt

import statsmodels.api as sm
from statkraft.ssa.wrappers import ReadWrapper

#Innlesning fra SMG
tz = pytz.timezone('Etc/GMT-1')
period = ReadWrapper(start_time='01.08.2018', end_time='01.01.2019', read_from='SMG_PROD', tz=tz)

names_y = ['/NVE.-Randsfjord....-D1004A0V', '/GLB.-Raudalsvatn...-D1004A5AI0101','/Roes-Roessvatn.....-D1004S0BT0132']
name_x = ['/Maar-Maarvatn......-D1004S0BT0132'] #/Norg-No1.Fasit.....-U9100S0BT0105
df_y = period.read(names_y)[:-100]
df_x = period.read(name_x)[:-100]
```

#### Her utfÃ¸res regresjonen:

```python
first_model = sm.OLS(df_x[name_x], df_y[names_y])
results = first_model.fit()
print(results.summary())
```

```python
results.pvalues
```

```python
plt.figure(figsize=(16, 10))
plt.plot(df_x[-100:], color='blue', label='fasit')
plt.plot(results.predict(df_y[-100:]), color='orange', label='regresjonsmodell')
plt.title('Resultat')
plt.legend()
```
